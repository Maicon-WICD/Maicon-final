```python
test1_path = './results/ee1'
test2_path = './results/3000pth_norm'
test3_path = './results/prenet2000_shift'
test4_path = './results/Denoised_Inference_230'
test5_path = './results/final_drop'
save_path = './results/final2'
file_list = os.listdir(test1_path)

for name in file_list:
    print(name)
    test1 = cv2.imread(os.path.join(test1_path, name))
    test2 = cv2.imread(os.path.join(test2_path, name))
    test3 = cv2.imread(os.path.join(test3_path, name))
    test4 = cv2.imread(os.path.join(test4_path, name))
    test5 = cv2.imread(os.path.join(test5_path, name))
    test1 = test1 // 5
    test2 = test2 // 5
    test3 = test3 // 5
    test4 = test4 // 5
    test5 = test5 // 5

    test = test1 + test2 + test3 + test4 + test5
    cv2.imwrite(os.path.join(save_path, name), test)


```
